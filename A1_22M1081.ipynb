# Linear Regression from Scratch

# Task-1:

A data matrix is basically a 2D array or a matrix in which each row and column represent some numerical or catogorical data. So in this question i have to generate a 2D array. Feature dimension is basically number of input variable.


```python
#impoting libraries

import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
samples = int(input('Enter the value of samples:'))   # taking no. of samples(N) from user
f_dim = int(input('Enter the value of dimension:'))   # taking no. of feature dimension(d) from user
def data_matrix_X(samples, f_dim):                    # defining function to generate data matrix(x_d) with input samples and f_dim
    x = random.randint(2, 6, size = (samples, f_dim)) # gerating 2D array of size (samples, f_dim) with random entries between 1 and 5 
    return x
    
x_d = data_matrix_X(samples, f_dim)                   # calling function data_matrix_x
print(x_d)
```

# Task-2:

I have generated data matrix of sample size = samples, and feature dimention = f_dim, as shown above. Now i will take the dot product of data matrix with randomly generated weight vector w then add bias and noise.

```python
def dep_var_t(x_d, w, w_0, n_var):            # function to calculate dependent variable t with inputs data matrix, weight vector, bias and noise variance 
    w_00 = w_0*np.ones(samples)               # array of bias with size = (N, 1) or bias vector
    noise = random.normal(loc = 0, scale = n_var, size = (samples))
    #noise = np.random.randn(samples)         # noise vector of size = (N, 1) 
    tar_vec_t = np.dot(x_d, w) + w_00 + noise # calculates target vector t = transpose(t)*x_d + bias + nosie
    return tar_vec_t                          # returns the target vector t

x_d = data_matrix_X(samples, f_dim)           # storing data matrix in x_d
w = random.randint(2, 8, size = (f_dim))      # generating weight vector of size = (d, 1) having random entries between 2 and 7 
print('Target variable:\n', dep_var_t(x_d, w, 5, 2)) # prints target variable

```

# Task-3:
The linear regression estimate is basically y = w0 + w1x1 + w2x2 +... which is basically dot product of the data matrix with weight vector.

```python
def lin_reg_est(x_d, w):       # function for linear regression estimate with input as data matrix(x_d) and weight vector(w)
    w_00 = 5*np.ones(samples)  # bias vector w00 (fixed) 
    y = np.dot(x_d, w) + w_00  # prediction of y
    return y                   # returns linear regression estimate y

print(lin_reg_est(x_d, w))
```

# Task-4:


```python
def mse():       # function to calculate MSE
    y = lin_reg_est(x_d, w)     # predicted value 
    t = dep_var_t(x_d, w, 5, 2) # actual value
    sqr_err = 0                 
    for i in range(len(y)):     # looping through length of predicted value
        sqr_err += pow((y[i] - t[i]), 2) # square of error
    m_sqr_err = sqr_err/(2*samples)      # mean of square error
    return m_sqr_err                     # return MSE
                           
print('Mean Square Error:\n', mse())
        
    
```

# Task-5:
Weights of the linear regression have been estimated using formula: Inv{(lamda* I) + Trans(phi) * phi)} * Trans(phi) * t.
#### Dimension of  Trans(phi)*phi is (DxN)(NxD) = (DxD) 
D = feature dimension = f_dim

phi = feature matrix, but here we have taken phi = data matrix (x_d)

```python
def w_est(x_d, t, lamda):       # function to estimate weight vector            
    I_mat = np.identity(f_dim)                    # identity matrix of size = (f_dim, f_dim)
    mat_1 = (lamda*I_mat) + np.dot(x_d.T, x_d)    # mat_1 = lamda*I + transpose(X)*X
    mat_inv = np.linalg.inv(mat_1)                # inverse of mat_1
    mat_2 = np.dot(mat_inv, x_d.T)                # mat_inv*transpose(X)
    w_estimate = np.dot(mat_2, t)                 # estimated weight = mat_2*t
    return w_estimate                             # returns estimated weight
lamda = float(input('Enter the value of lamda:')) # taking lamda value from user
t = dep_var_t(x_d, w, 5, 2)
print(w_est(x_d, t, lamda))
```

# Task-6:
To compute the gradient of MSE we have to do the following:

1) Square of the error is square(yi - ti) so (yi - ti) gives the error in sample point.

2) Take the partial derivative of the loss function w.r.t wj and after solving and rearranging the we get the expression of gradient vector which is 
### dL/dWj = (1/N) * (e1x1j + e2x2j + e3x3j + ....),  where ei = (yi - ti) and xij data sample point.

```python
def grad_mse(x_d, t, w):       # function for gradient of MSE  
    y_i = lin_reg_est(x_d, w)  # calling function y (linear regression estimate) 
    e_i = y_i - t              # substractig vector y and vector t  
    grad_ms_err = np.zeros(len(w))  # array of size = (d, 1) 
    for j in range(len(w)):         # looping through lenght of weight vector
        x_ij = np.zeros(len(x_d))   # array of zeros to store features 
        for k in range(len(x_d)):   # looping through the columns of data matrix 
            x_ij[k] += x_d[k, j]    # extracting the feature of data matrix(column of x_d)
        grad_ms_err[j] = (1/samples)*np.dot(x_ij, e_i) # divide by number of samples
    return grad_ms_err              # returns gradient of MSE

w = random.randint(20, 80, size = (f_dim))
x_d = data_matrix_X(samples, f_dim)
t = dep_var_t(x_d, w, 5, 2)
print(grad_mse(x_d, t, w))
        
```

# Task-7:
L2 norm of a vector is basically sum of square of mod value of each element in vector.

```python
def L2_norm(w):               # function for L2 norm with input as weight vector
    L2 = 0              
    for i in range(len(w)):   # iterating all values of weight vector  
        L2 += np.square(w[i]) # taking square of values
    return L2                 # returns L2 norm


print(L2_norm(w))
```

# Task-8:
Gradient of L2 norm = 2|w|

```python
def grad_L2_norm(w):  # function to calculate gradient of L2 norm with input weight vector
    norm = 2*w        # gradient of L2 norm is 2*W 
    return norm       # returns gradien of L2 norm

print(w)
print('Gradient of L2 norm:\n', grad_L2_norm(w))
```

# Task-9:
L1 norm is sum of the absolute weight vector values. 

```python
def L1_norm(w):             # funtion for L1 norm with weight vector
    absl = np.absolute(w)   # taking absolute of all weights 
    L1 = 0
    for i in range(len(w)): # going through all values
        L1 += absl[i]       # adding all absolute weight values
    
    return L1               # returns L1 norm


print(w)

print(L1_norm(w))
```

# Task-10:
Gradient of L1 norm = 1 w.r.t any weight w, therefore Gradient vector is array of ones having size = f_dim x 1

```python
def grad_L1_norm(w):           # function to calculate gradient of L1 norm with given weight vector 
    grad_L1 = np.zeros(len(w)) # array of zeros of lenght = weight vector lenght
    for i in range(len(w)):    # iterating through all weights
        if w[i] > 0:           # checking the sign of weight
            grad_L1[i] = 1     # assign 1 if weight is positive 
        else:                  
            grad_L1[i] = -1    # assign -1 if weigh is negative
    return grad_L1             # returns gradient of L1 norm

print(grad_L1_norm(w))
```

# Task-11:
In this task i have taken all 3 loss function which is MSE, L2 norm, and L1 norm. So total cost funtion is equal to:
### L(W) = MSE + L2 + L1
Now i will calculate the gradient of above L(w) which is given as:
### dL/dWj = grad(MSE) + grad(L2) + grad(L1)
### dL/dWj = (1/N) * (e1x1j + e2x2j + e3x3j + ....) + (lamda_2)Wj + lamda_1
### where ei = (yi - ti), xij = data sample point. lamda_2 and  lamda_1 are some positive constant.



```python
def update_w_linreg(x_d, t, w, eta, lamd_2, lamd_1): # defining function to update of weights of linear regression using gradient descent 
    grad_MSE = grad_mse(x_d, t, w)           # calling function grad_mse and it gives gradient of MSE
    grad_L2 = grad_L2_norm(w)                # calling function grad_L2_norm and it gives gradient of L2 norm
    grad_L1 = lamd_1*grad_L1_norm(w)         # calling function grad_L1_norm and it gives gradient of L1 norm
    delta_L =  grad_MSE + grad_L2 + grad_L1  # adds all norms
    
    w = w - (eta*(delta_L))                  # updating weights by subtracting step size(learning_rate*gradient_of_cost funuction)
    mse_up = mse()                           # updates MSE based on new weight vector 
    cheers = 'Cheers!!'                      # little happiness
    return w, mse_up, cheers                 # returns updated weight vector, updated MSE and happiness  

print('With L1 = 0:', update_w_linreg(x_d, t, w, 0.01, 2, 0))    # with L1 norm = 0
#print('With L2 = 0:', update_w_linreg(x_d, t, w, 0.01, 0, 1.5)) # with L2 norm = 0   
```
